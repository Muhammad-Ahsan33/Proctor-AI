```python
# ── Core imports ──────────────────────────────────────────────────────────
import os, json, time, re, warnings
import numpy as np
import cv2
from PIL import Image
import torch
warnings.filterwarnings('ignore')

# NLP
from sentence_transformers import SentenceTransformer, util

# OCR
import easyocr

# Whisper
import whisper

# MediaPipe
# import mediapipe as mp

# AI content detection
from transformers import pipeline as hf_pipeline

# Gradio
import gradio as gr

print('✅ All imports successful.')
```

```python
# ── Load shared models (once) ─────────────────────────────────────────────
print('Loading NLP model...')
nlp_model = SentenceTransformer('all-MiniLM-L6-v2')

print('Loading OCR reader...')
ocr_reader = easyocr.Reader(['en'], gpu=torch.cuda.is_available())

print('Loading Whisper model...')
whisper_model = whisper.load_model('base')

print('Loading AI content detector...')
ai_detector = hf_pipeline(
    'text-classification',
    model='roberta-base-openai-detector',
    device=0 if torch.cuda.is_available() else -1
)

print('✅ All models loaded.')
```

```python
# ── Shared utility functions ──────────────────────────────────────────────

# Global results store (in-memory for this session)
RESULTS_DB = {}

def compute_similarity(student_answer: str, model_answer: str) -> float:
    """Semantic similarity score between 0 and 1."""
    if not student_answer.strip():
        return 0.0
    emb1 = nlp_model.encode(student_answer, convert_to_tensor=True)
    emb2 = nlp_model.encode(model_answer, convert_to_tensor=True)
    score = float(util.cos_sim(emb1, emb2)[0][0])
    return round(max(0.0, min(1.0, score)), 3)

def score_answers(student_answers: dict, answer_key: dict) -> dict:
    """
    Score student answers against answer key.
    student_answers: {q_id: student_text}
    answer_key:      {q_id: {question, answer, marks}}
    Returns: {q_id: {score, max_marks, similarity, feedback}}
    """
    results = {}
    total_scored = 0
    total_marks  = 0
    for q_id, meta in answer_key.items():
        student_ans = student_answers.get(q_id, '')
        sim = compute_similarity(student_ans, meta['answer'])
        max_marks = meta.get('marks', 10)
        scored = round(sim * max_marks, 2)
        total_scored += scored
        total_marks  += max_marks
        if sim >= 0.8:
            feedback = 'Excellent answer.'
        elif sim >= 0.6:
            feedback = 'Good but could be more complete.'
        elif sim >= 0.4:
            feedback = 'Partially correct. Review the concept.'
        else:
            feedback = 'Incorrect or very incomplete.'
        results[q_id] = {
            'question'   : meta['question'],
            'student_ans': student_ans,
            'model_ans'  : meta['answer'],
            'similarity' : sim,
            'score'      : scored,
            'max_marks'  : max_marks,
            'feedback'   : feedback
        }
    results['__total__'] = {'scored': round(total_scored, 2), 'max': total_marks}
    return results

def print_score_report(results: dict, test_type: str, student_name: str):
    """Pretty-print result to notebook output."""
    total = results.pop('__total__', {'scored': 0, 'max': 0})
    print(f'\n{'='*60}')
    print(f'  📄 {test_type} — Result for: {student_name}')
    print(f'{'='*60}')
    for q_id, r in results.items():
        print(f'\n  Q{q_id}: {r["question"]}')
        print(f'  Student : {r["student_ans"] or "(no answer)"}')
        print(f'  Expected: {r["model_ans"]}')
        print(f'  Score   : {r["score"]} / {r["max_marks"]}  (similarity: {r["similarity"]})')
        print(f'  {r["feedback"]}')
    results['__total__'] = total
    print(f'\n  🏆 TOTAL : {total["scored"]} / {total["max"]}')
    print(f'{'='*60}\n')

print('✅ Utility functions ready.')
```

```python
# ── 2A: Image preprocessing (light cleanup before sending to vision model) ─
# ── Improved Preprocessing for Handwriting ───────────────────────────────
def preprocess_image(image_path: str, out_path: str = '/tmp/processed_exam.jpg') -> str:
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Cannot read image: {image_path}")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Stronger preprocessing for handwriting
    denoised = cv2.fastNlMeansDenoising(gray, h=15)
    blur = cv2.GaussianBlur(denoised, (3,3), 0)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    enhanced = clahe.apply(blur)

    # Binarization (helps a lot with handwriting)
    _, binary = cv2.threshold(enhanced, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    enhanced_rgb = cv2.cvtColor(binary, cv2.COLOR_GRAY2BGR)
    cv2.imwrite(out_path, enhanced_rgb)
    print(f'  Preprocessed (binarized) image saved → {out_path}')
    return out_path

print('✅ Preprocessing function ready.')
```

```python
def extract_text_from_image(image_path: str) -> str:
    print('  Preprocessing image...')
    processed_path = preprocess_image(image_path)

    print('  Running OCR with improved settings...')
    # Better settings for handwriting
    results = ocr_reader.readtext(
        processed_path,
        detail=0,
        paragraph=True,
        width_ths=0.7,          # helps with word spacing
        height_ths=0.7,
        text_threshold=0.6
    )
    full_text = ' '.join(results)
    print(f'  Extracted {len(full_text)} characters.')
    return full_text
```

```python
def parse_handwritten_answers(raw_text: str, exam: dict) -> dict:
    """
    Attempt to split OCR text by question numbers.
    Looks for patterns like Q1, Q2, 1., 2., Answer 1, etc.
    Falls back to splitting text equally if no markers found.
    """
    answers = {}
    # Try to find question markers
    pattern = r'(?:Q|Question|Ans(?:wer)?)?\s*([0-9]+)[\.\.:\)]'
    splits = re.split(pattern, raw_text, flags=re.IGNORECASE)

    # splits alternates: [pre, num, text, num, text, ...]
    i = 1
    found_any = False
    while i < len(splits) - 1:
        q_num = splits[i].strip()
        q_text = splits[i + 1].strip() if i + 1 < len(splits) else ''
        if q_num in exam:
            answers[q_num] = q_text
            found_any = True
        i += 2

    # Fallback: split entire text equally across questions
    if not found_any:
        q_ids = list(exam.keys())
        chunk = max(1, len(raw_text) // len(q_ids))
        for idx, q_id in enumerate(q_ids):
            answers[q_id] = raw_text[idx * chunk: (idx + 1) * chunk].strip()

    # Fill missing questions with empty string
    for q_id in exam:
        if q_id not in answers:
            answers[q_id] = ''
    return answers

print('✅ Parser ready.')
```

```python
import ollama

def extract_text_from_image_llama(image_path: str) -> str:
    print("  🔍 Using Llama 3.2 Vision for OCR...")
    prompt = """Extract the handwritten text from this image exactly as written.
Return only the raw text. Preserve any question or answer numbers like Q1."""

    response = ollama.chat(
        model='llama3.2-vision',
        messages=[{
            'role': 'user',
            'content': prompt,
            'images': [image_path]
        }]
    )
    text = response['message']['content'].strip()
    print(f"  ✅ Extracted {len(text)} characters.")
    return text
```

```python
def score_answers(student_answers: dict, answer_key: dict) -> dict:
    results = {}
    total_scored = 0
    total_marks = 0

    for q_id, meta in answer_key.items():
        student_ans = student_answers.get(q_id, '').strip()
        model_ans = meta['answer']
        max_marks = meta.get('marks', 10)

        student_clean = student_ans.lower()

        if not student_ans or student_clean in ["(no answer)", "no answer", "n/a", "-", "none"]:
            scored = 0
            feedback = "❌ No answer"

        else:
            sim = compute_similarity(student_ans, model_ans)

            # Much more lenient scoring
            if sim >= 0.5:
                scored = max_marks
                feedback = "✅ Correct concept"
            elif sim >= 0.3:
                scored = round(max_marks * 0.7, 1)
                feedback = "🟡 Partially correct"
            else:
                # Keyword bonus
                keywords = {"machine", "learning", "artificial", "intelligence", "learn", "data", "experience", "programmed"}
                student_lower = student_ans.lower()
                matches = sum(1 for kw in keywords if kw in student_lower)
                if matches >= 3:
                    scored = round(max_marks * 0.6, 1)
                    feedback = "🟠 Has key concepts"
                else:
                    scored = round(max_marks * 0.2, 1)
                    feedback = "❌ Mostly incorrect"

        total_scored += scored
        total_marks += max_marks

        results[q_id] = {
            'question': meta['question'],
            'student_ans': student_ans or "(no answer)",
            'model_ans': model_ans,
            'similarity': round(sim, 3),
            'score': scored,
            'max_marks': max_marks,
            'feedback': feedback
        }

    results['__total__'] = {'scored': round(total_scored, 2), 'max': total_marks}
    return results
```

# ⌨️ SECTION 3 — Typing Test
> Student types answers in a Gradio widget. System captures keystrokes, detects cheating, and scores answers.

```python
# ── 3A: Copy-paste detection (Python-side, text-delta approach) ───────────
#
# How it works:
#   Every time a Gradio textbox fires a 'change' event we receive the
#   full current text. We diff it against the previous snapshot.
#   Normal typing adds 1-2 chars per event.
#   A paste adds tens or hundreds of chars in a single event → flagged.
#
# PASTE_THRESHOLD: minimum characters added in one event to count as paste.
PASTE_THRESHOLD = 20

def analyze_text_deltas(delta_log: list) -> dict:
    """
    delta_log: list of int — character-count change per change event.
    Returns cheating report dict.
    """
    paste_events = [d for d in delta_log if d >= PASTE_THRESHOLD]
    report = {
        'copy_paste_detected': len(paste_events) > 0,
        'paste_event_count'  : len(paste_events),
        'largest_paste'      : max(paste_events, default=0),
        'details'            : [
            f'Paste event: +{d} characters inserted at once'
            for d in paste_events
        ]
    }
    return report

print(f'✅ Copy-paste detector ready. Threshold = {PASTE_THRESHOLD} chars per event.')
```

```python
# ── 3B: AI-generated content detection ───────────────────────────────────
def detect_ai_content(text: str) -> dict:
    """Returns whether text appears AI-generated and confidence score."""
    if not text.strip() or len(text.split()) < 10:
        return {'ai_generated': False, 'confidence': 0.0,
                'label': 'Too short to analyze'}
    try:
        result = ai_detector(text[:512])[0]
        label  = result['label']   # 'LABEL_1' = AI, 'LABEL_0' = Human
        score  = round(result['score'], 3)
        is_ai  = (label == 'LABEL_1') and (score > 0.7)
        return {
            'ai_generated': is_ai,
            'confidence'  : score,
            'label'       : f"{'AI-generated' if is_ai else 'Human-written'} ({score*100:.1f}% confidence)"
        }
    except Exception as e:
        return {'ai_generated': False, 'confidence': 0.0,
                'label': f'Detection error: {e}'}

print('✅ AI content detector ready.')
```

```python
# ── 3C: Gradio Typing Test Widget ─────────────────────────────────────────
#
# Design:
#   • Each answer textbox fires .change() on every keystroke.
#   • A hidden gr.State per question stores a list of character-count deltas.
#   • On each change event: delta = len(new_text) - len(prev_text).
#     If delta >= PASTE_THRESHOLD  → paste detected for that question.
#   • On Submit: deltas are analysed, AI detection runs, answers are scored.

Q_LIST = [(q_id, meta['question']) for q_id, meta in TYPING_EXAM.items()]
NUM_Q  = len(Q_LIST)

with gr.Blocks(title='Typing Test') as typing_app:

    gr.Markdown(
        '# ⌨️ AI Examination — Typing Test\n'
        '> Type your answers below. **Do not copy-paste.** '
        'All keystrokes are monitored.'
    )

    student_name_box = gr.Textbox(
        label='Your Full Name',
        placeholder='e.g. User'
    )

    answer_boxes  = []   # gr.Textbox per question
    prev_states   = []   # gr.State — previous text snapshot
    delta_states  = []   # gr.State — list of deltas so far
    live_flags    = []   # gr.Markdown — live warning per question

    for q_id, question in Q_LIST:
        gr.Markdown(f'---\n### Q{q_id}: {question}')

        ans_box   = gr.Textbox(
            label=f'Answer {q_id}',
            lines=4,
            placeholder='Type your answer here (do NOT paste)...'
        )
        prev_st   = gr.State('')          # previous text value
        delta_st  = gr.State([])          # running list of deltas
        flag_md   = gr.Markdown('')       # live warning shown to student

        answer_boxes.append(ans_box)
        prev_states.append(prev_st)
        delta_states.append(delta_st)
        live_flags.append(flag_md)

        # ── Live change handler (fires on every keystroke) ──
        # We use a closure to capture q_id for the warning message.
        def make_change_handler(qid):
            def on_change(new_text, prev_text, deltas):
                delta = len(new_text) - len(prev_text)
                new_deltas = deltas + [delta]
                # Warn student immediately if paste detected
                if delta >= PASTE_THRESHOLD:
                    warning = (
                        f'⚠️ **WARNING Q{qid}**: Large text insertion detected '
                        f'(+{delta} chars). This has been logged as a possible paste.'
                    )
                else:
                    warning = ''
                return new_text, new_deltas, warning
            return on_change

        ans_box.change(
            fn=make_change_handler(q_id),
            inputs=[ans_box, prev_st, delta_st],
            outputs=[prev_st, delta_st, flag_md]
        )

    gr.Markdown('---')
    submit_btn = gr.Button('📤 Submit Test', variant='primary', size='lg')
    result_out = gr.Markdown('')

    # ── Submit handler ──
    def process_submission(student_name, *args):
        """
        args order: [answers × N] + [delta_lists × N]
        """
        if not student_name.strip():
            return '⚠️ Please enter your name first.'

        answers     = list(args[:NUM_Q])
        delta_lists = list(args[NUM_Q:])

        student_answers   = {}
        paste_reports     = {}
        ai_reports        = {}

        for i, (q_id, _) in enumerate(Q_LIST):
            text = answers[i] if i < len(answers) else ''
            deltas = delta_lists[i] if i < len(delta_lists) else []
            student_answers[q_id] = text
            paste_reports[q_id]   = analyze_text_deltas(deltas)
            ai_reports[q_id]      = detect_ai_content(text)

        # Score
        score_report = score_answers(student_answers, TYPING_EXAM)
        total = score_report.get('__total__', {'scored': 0, 'max': 0})

        # Aggregate cheating
        any_paste = any(r['copy_paste_detected'] for r in paste_reports.values())
        any_ai    = any(r['ai_generated']        for r in ai_reports.values())
        details   = []
        for q_id in student_answers:
            pr = paste_reports[q_id]
            ar = ai_reports[q_id]
            if pr['copy_paste_detected']:
                details.append(
                    f'Q{q_id}: {pr["paste_event_count"]} paste event(s) — '
                    f'largest was +{pr["largest_paste"]} chars'
                )
                for d in pr['details']:
                    details.append(f'  └ {d}')
            if ar['ai_generated']:
                details.append(f'Q{q_id}: AI-generated content — {ar["label"]}')

        cheating_report = {
            'copy_paste_detected': any_paste,
            'ai_content_detected': any_ai,
            'details'            : details
        }

        # Persist to RESULTS_DB
        if student_name not in RESULTS_DB:
            RESULTS_DB[student_name] = {}
        RESULTS_DB[student_name]['Typing Test'] = {
            'score_report'   : score_report,
            'cheating_report': cheating_report
        }

        # ── Build result markdown ──
        out  = f'## ✅ Submitted — {student_name}\n\n'
        out += f'### 🏆 Score: {total["scored"]} / {total["max"]}\n\n'
        out += '### 📋 Per-question breakdown\n'
        for q_id, r in score_report.items():
            if q_id == '__total__':
                continue
            out += f'- **Q{q_id}**: {r["score"]} / {r["max_marks"]} — {r["feedback"]}\n'

        out += '\n### 🔍 Cheating Report\n'
        out += ('- ⚠️ **Copy-paste DETECTED**\n' if any_paste
                else '- ✅ No copy-paste detected\n')
        out += ('- ⚠️ **AI-generated content DETECTED**\n' if any_ai
                else '- ✅ No AI-generated content detected\n')

        if details:
            out += '\n**Evidence:**\n'
            for d in details:
                out += f'  - {d}\n'

        return out

    submit_btn.click(
        fn=process_submission,
        inputs=[student_name_box] + answer_boxes + delta_states,
        outputs=result_out
    )

print('✅ Typing test widget built. Run next cell to launch.')
```

```python
# ── 4A: Audio extraction + Whisper transcription ─────────────────────────
def extract_audio_from_video(video_path: str, audio_out: str = '/tmp/exam_audio.wav') -> str:
    """Extract audio track from video file."""
    import moviepy.editor as mpy
    clip = mpy.VideoFileClip(video_path)
    clip.audio.write_audiofile(audio_out, verbose=False, logger=None)
    print(f'  Audio extracted → {audio_out}')
    return audio_out

def transcribe_audio(audio_path: str) -> str:
    """Transcribe audio to text using Whisper."""
    print('  Transcribing audio with Whisper...')
    result = whisper_model.transcribe(audio_path)
    text = result['text'].strip()
    print(f'  Transcript: {text[:100]}...' if len(text) > 100 else f'  Transcript: {text}')
    return text

def parse_spoken_answers(transcript: str, exam: dict) -> dict:
    """
    Split transcript by question markers spoken aloud.
    Looks for 'question one', 'question 1', 'answer one', etc.
    Falls back to splitting equally.
    """
    answers = {}
    num_words = {
        '1':'one','2':'two','3':'three','4':'four','5':'five',
        '6':'six','7':'seven','8':'eight','9':'nine','10':'ten'
    }
    lower = transcript.lower()
    breakpoints = {}
    for q_id, word in num_words.items():
        if q_id not in exam:
            continue
        for pattern in [
            f'question {word}', f'question {q_id}',
            f'answer {word}',   f'answer {q_id}',
            f'q{q_id}'
        ]:
            idx = lower.find(pattern)
            if idx != -1:
                breakpoints[q_id] = idx
                break

    if len(breakpoints) >= 2:
        sorted_bps = sorted(breakpoints.items(), key=lambda x: x[1])
        for i, (q_id, start) in enumerate(sorted_bps):
            end = sorted_bps[i + 1][1] if i + 1 < len(sorted_bps) else len(transcript)
            answers[q_id] = transcript[start:end].strip()
    else:
        # Fallback: split equally
        q_ids = list(exam.keys())
        chunk = max(1, len(transcript) // len(q_ids))
        for idx, q_id in enumerate(q_ids):
            answers[q_id] = transcript[idx * chunk: (idx + 1) * chunk].strip()

    for q_id in exam:
        if q_id not in answers:
            answers[q_id] = ''
    return answers

print('✅ Audio & transcription functions ready.')
```

```python
from ultralytics import YOLO
yolo_model = YOLO('yolov8n.pt')  # small fast model
PHONE_CLASS_ID = 67               # COCO class id for 'cell phone'

def detect_phone_in_frame(frame):
    results = yolo_model(frame, verbose=False)[0]
    for box in results.boxes:
        if int(box.cls[0]) == PHONE_CLASS_ID and float(box.conf[0]) > 0.4:
            return True
    return False
```

```python
import librosa
import numpy as np

def analyze_audio_proctoring(audio_path: str):
    y, sr = librosa.load(audio_path, sr=16000)
    events = []

    # --- Whisper volume spike (loud noise, no speech) ---
    frame_length = sr // 2  # 0.5s chunks
    for i in range(0, len(y) - frame_length, frame_length):
        chunk = y[i:i+frame_length]
        rms = np.sqrt(np.mean(chunk**2))
        timestamp = round(i / sr, 1)
        if rms > 0.08:  # tune this threshold
            events.append((timestamp, "AUDIO_SPIKE", f"RMS:{rms:.3f}"))

    # --- Multiple speakers (pitch variation = different voices) ---
    pitches, _ = librosa.piptrack(y=y, sr=sr)
    active = pitches[pitches > 0]
    if len(active) > 0:
        pitch_std = np.std(active)
        if pitch_std > 800:  # high variation = multiple voices
            events.append((0.0, "MULTIPLE_SPEAKERS", f"Pitch std:{pitch_std:.0f}"))

    return {
        'audio_spike_events'     : len([e for e in events if e[1]=="AUDIO_SPIKE"]),
        'multiple_speakers'      : any(e[1]=="MULTIPLE_SPEAKERS" for e in events),
        'details'                : [f"[{t:.1f}s] {typ}: {msg}" for t,typ,msg in events]
    }
```

```python
# ====================== FINAL EXAM PROCTORING (Colab Ready) ======================
import cv2
import numpy as np
from collections import deque
import mediapipe as mp

print("🔧 Initializing Student Proctoring System...")

# ========================= CONFIG =========================
BUFFER_WINDOW = 12
look_down_buffer = deque(maxlen=BUFFER_WINDOW)
EAR_THRESHOLD = 0.25
MAR_THRESHOLD = 0.5
YAW_THRESHOLD = 35
PITCH_THRESHOLD = 25
FACE_COUNT_ALERT = 2
EMA_ALPHA = 0.35

# ========================= MEDIA PIPE SETUP =========================
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=5,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

# ========================= BUFFERS =========================
ear_buffer = deque(maxlen=BUFFER_WINDOW)
mar_buffer = deque(maxlen=BUFFER_WINDOW)
gaze_buffer = deque(maxlen=BUFFER_WINDOW)

_ema_pitch = None
_ema_yaw = None

# Landmarks
LEFT_EYE = [362, 385, 387, 263, 373, 380]
RIGHT_EYE = [33, 160, 158, 133, 153, 144]
MOUTH = [78, 308, 13, 14]

MODEL_POINTS = np.array([
    (0.0, 0.0, 0.0), (-225.0, 170.0, -135.0), (225.0, 170.0, -135.0),
    (-150.0, -150.0, -125.0), (150.0, -150.0, -125.0)
], dtype=np.float32)


# ========================= HELPERS =========================
def calculate_ear(landmarks, eye_indices, w, h):
    pts = [np.array([landmarks[i].x * w, landmarks[i].y * h]) for i in eye_indices]
    return (np.linalg.norm(pts[1]-pts[5]) + np.linalg.norm(pts[2]-pts[4])) / (2 * np.linalg.norm(pts[0]-pts[3]) + 1e-6)


def calculate_mar(landmarks, mouth_indices, w, h):
    pts = [np.array([landmarks[i].x * w, landmarks[i].y * h]) for i in mouth_indices]
    return np.linalg.norm(pts[2]-pts[3]) / (np.linalg.norm(pts[0]-pts[1]) + 1e-6)


def estimate_head_pose(landmarks, w, h):
    global _ema_pitch, _ema_yaw
    image_points = np.array([
        (landmarks[1].x*w, landmarks[1].y*h),
        (landmarks[152].x*w, landmarks[152].y*h),
        (landmarks[33].x*w, landmarks[33].y*h),
        (landmarks[263].x*w, landmarks[263].y*h),
        (landmarks[61].x*w, landmarks[61].y*h),
    ], dtype=np.float32)

    focal = w
    cam_matrix = np.array([[focal, 0, w/2], [0, focal, h/2], [0, 0, 1]], dtype=np.float32)

    success, rot_vec, _ = cv2.solvePnP(MODEL_POINTS, image_points, cam_matrix,
                                       np.zeros((4,1)), flags=cv2.SOLVEPNP_SQPNP)

    if not success:
        return _ema_pitch or 0.0, _ema_yaw or 0.0

    rot_mat, _ = cv2.Rodrigues(rot_vec)
    _, _, _, _, _, _, euler = cv2.decomposeProjectionMatrix(np.hstack((rot_mat, np.zeros((3,1)))))

    raw_pitch = float(euler[0][0])
    raw_yaw = -float(euler[1][0])

    if _ema_pitch is None:
        _ema_pitch, _ema_yaw = raw_pitch, raw_yaw
    else:
        _ema_pitch = EMA_ALPHA * raw_pitch + (1 - EMA_ALPHA) * _ema_pitch
        _ema_yaw = EMA_ALPHA * raw_yaw + (1 - EMA_ALPHA) * _ema_yaw

    return _ema_pitch, _ema_yaw



# ========================= MAIN FUNCTION =========================
def analyze_video_proctoring(video_path: str):
    global _ema_pitch, _ema_yaw
    _ema_pitch = _ema_yaw = None  # Reset for new video

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS) or 30
    frame_idx = 0
    events = []
    face_counts = []

    print(f"🎥 Analyzing video: {video_path}")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_idx += 1
        if frame_idx % 3 != 0:
            continue

        timestamp = round(frame_idx / fps, 2)
        h, w = frame.shape[:2]

        if frame_idx % 15 == 0:
          if detect_phone_in_frame(frame):
              events.append((timestamp, "PHONE_DETECTED", "Cell phone visible in frame"))

        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_mesh.process(rgb)

        current_faces = len(results.multi_face_landmarks) if results.multi_face_landmarks else 0
        face_counts.append(current_faces)

        is_distracted = is_yawning = is_drowsy = False

        if results.multi_face_landmarks:
            for face_landmarks in results.multi_face_landmarks:
                landmarks = face_landmarks.landmark

                # Face size = distance between eye corners (already have landmarks)
                left_corner  = np.array([landmarks[33].x * w,  landmarks[33].y * h])
                right_corner = np.array([landmarks[263].x * w, landmarks[263].y * h])
                face_width = np.linalg.norm(right_corner - left_corner)

                LEAN_THRESHOLD = w * 0.45  # face takes up >45% of frame width
                if face_width > LEAN_THRESHOLD:
                    events.append((timestamp, "TOO_CLOSE", f"Face width:{face_width:.0f}px"))

                # Gaze Detection
                pitch, yaw = estimate_head_pose(landmarks, w, h)
                # Pitch > 0 means looking down in your coordinate system
                LOOK_DOWN_PITCH = -35
                look_down_buffer.append(pitch < -LOOK_DOWN_PITCH)
                if sum(look_down_buffer) >= int(0.9 * BUFFER_WINDOW):
                    events.append((timestamp, "EYES_DOWN", f"Pitch:{pitch:.1f}°"))

                distracted = abs(yaw) > YAW_THRESHOLD or abs(pitch) > PITCH_THRESHOLD

                gaze_buffer.append(distracted)
                if sum(gaze_buffer) >= int(0.7 * BUFFER_WINDOW):
                    is_distracted = True

                # EAR & MAR
                ear = (calculate_ear(landmarks, LEFT_EYE, w, h) +
                       calculate_ear(landmarks, RIGHT_EYE, w, h)) / 2
                mar = calculate_mar(landmarks, MOUTH, w, h)

                ear_buffer.append(ear < EAR_THRESHOLD)
                mar_buffer.append(mar > MAR_THRESHOLD)

                if sum(ear_buffer) >= int(0.8 * BUFFER_WINDOW):
                    is_drowsy = True
                if sum(mar_buffer) >= int(0.6 * BUFFER_WINDOW):
                    is_yawning = True

        else:
            gaze_buffer.append(True)

        # Log suspicious events
        if current_faces == 0:
            events.append((timestamp, "NO_FACE", "Student not visible"))
        elif current_faces >= FACE_COUNT_ALERT:
            events.append((timestamp, "MULTIPLE_FACES", f"{current_faces} people in frame"))

        if is_distracted:
            events.append((timestamp, "GAZE_AWAY", f"Yaw:{yaw:.1f}° Pitch:{pitch:.1f}°"))
        if is_yawning:
            events.append((timestamp, "YAWNING", ""))
        if is_drowsy:
            events.append((timestamp, "DROWSY", ""))

    cap.release()

    # Summary
    avg_faces = np.mean(face_counts) if face_counts else 0
    return {
    'cheating_detected'   : avg_faces < 0.7 or any(c >= FACE_COUNT_ALERT for c in face_counts)
                            or len([e for e in events if e[1]=="GAZE_AWAY"]) > 20,
    'avg_faces'           : round(avg_faces, 2),
    'no_face_events'      : len([e for e in events if e[1]=="NO_FACE"]),
    'multiple_face_events': len([e for e in events if e[1]=="MULTIPLE_FACES"]),
    'gaze_away_events'    : len([e for e in events if e[1]=="GAZE_AWAY"]),
    'eyes_down_events'    : len([e for e in events if e[1]=="EYES_DOWN"]),     # new
    'too_close_events'    : len([e for e in events if e[1]=="TOO_CLOSE"]),     # new
    'phone_detected'      : any(e[1]=="PHONE_DETECTED" for e in events),       # new
    'details'             : [f"[{t:.1f}s] {typ}: {msg}" for t,typ,msg in events[:30]]
}

```

```python
# ── 4C: Run Online Meeting Test ───────────────────────────────────────────
STUDENT_NAME_OL = 'User'
VIDEO_PATH      = '/content/video3.mp4'

print(f'\n🎥 Running Online Meeting Test for: {STUDENT_NAME_OL}')
print('─' * 50)

# Step 1: Transcribe audio
print('\n[Step 1] Audio transcription...')
try:
    audio_path = extract_audio_from_video(VIDEO_PATH)
    transcript = transcribe_audio(audio_path)
except Exception as e:
    print(f'  Audio extraction failed ({e})')
    audio_path = None
    transcript = ''

spoken_answers = parse_spoken_answers(transcript, ONLINE_EXAM)
score_report   = score_answers(spoken_answers, ONLINE_EXAM)

# Step 2: CV proctoring
print('\n[Step 2] Running Advanced Proctoring...')
cheating_report = analyze_video_proctoring(VIDEO_PATH)

# Step 3: Audio proctoring
print('\n[Step 3] Running Audio Proctoring...')
if audio_path:
    audio_report = analyze_audio_proctoring(audio_path)
    cheating_report['audio_spike_events'] = audio_report['audio_spike_events']
    cheating_report['multiple_speakers']  = audio_report['multiple_speakers']
    cheating_report['details']           += audio_report['details']
    if audio_report['multiple_speakers'] or audio_report['audio_spike_events'] > 5:
        cheating_report['cheating_detected'] = True
else:
    cheating_report['audio_spike_events'] = 0
    cheating_report['multiple_speakers']  = False

# Step 4: Save results
if STUDENT_NAME_OL not in RESULTS_DB:
    RESULTS_DB[STUDENT_NAME_OL] = {}
RESULTS_DB[STUDENT_NAME_OL]['Online Test'] = {
    'score_report'   : score_report,
    'cheating_report': cheating_report
}

# Step 5: Print score
print_score_report(score_report, 'Online Meeting Test', STUDENT_NAME_OL)

# Step 6: Print full proctoring report
print('\n🔍 Proctoring Report:')
print(f'  Cheating detected    : {cheating_report["cheating_detected"]}')
print(f'  Average faces        : {cheating_report["avg_faces"]}')
print(f'  No-face frames       : {cheating_report["no_face_events"]}')
print(f'  Multiple faces       : {cheating_report["multiple_face_events"]}')
print(f'  Gaze away events     : {cheating_report["gaze_away_events"]}')
print(f'  Eyes down events     : {cheating_report.get("eyes_down_events", 0)}')
print(f'  Too close events     : {cheating_report.get("too_close_events", 0)}')
print(f'  Phone detected       : {cheating_report.get("phone_detected", False)}')
print(f'  Audio spikes         : {cheating_report.get("audio_spike_events", 0)}')
print(f'  Multiple speakers    : {cheating_report.get("multiple_speakers", False)}')

if cheating_report.get('details'):
    print('\n  Timeline of events:')
    for d in cheating_report['details']:
        print(f'    • {d}')
```

```python
# ── 5A: Combined summary report ───────────────────────────────────────────
def print_combined_report():
    if not RESULTS_DB:
        print('No test results found. Run tests in sections 2, 3, and 4 first.')
        return

    print('\n' + '='*70)
    print('         📊 COMBINED EXAMINATION REPORT — ALL STUDENTS')
    print('='*70)

    for student, tests in RESULTS_DB.items():
        print(f'\n  👤 Student: {student}')
        print('  ' + '─'*60)

        grand_scored = 0
        grand_max    = 0
        any_cheating = False

        for test_type, data in tests.items():
            sr = data.get('score_report', {})
            cr = data.get('cheating_report', {})
            total = sr.get('__total__', {'scored': 0, 'max': 0})

            grand_scored += total.get('scored', 0)
            grand_max    += total.get('max', 0)

            cheating_flags = []
            if cr.get('copy_paste_detected'):    cheating_flags.append('Copy-paste')
            if cr.get('ai_content_detected'):    cheating_flags.append('AI content')
            if cr.get('multiple_face_events', 0) > 2: cheating_flags.append('Multiple faces')
            if cr.get('no_face_events', 0) > 5:       cheating_flags.append('Absent from camera')
            if cr.get('head_pose_events', 0) > 10:    cheating_flags.append('Repeated distraction')
            if cr.get('eye_away_events', 0) > 10:     cheating_flags.append('Looking away')

            if cheating_flags:
                any_cheating = True

            score_pct = round((total.get('scored', 0) / max(total.get('max', 1), 1)) * 100, 1)
            status = '⚠️ CHEATING FLAGGED' if cheating_flags else '✅ Clean'

            print(f'\n  📝 {test_type}')
            print(f'     Score    : {total.get("scored",0)} / {total.get("max",0)} ({score_pct}%)')
            print(f'     Integrity: {status}')
            if cheating_flags:
                print(f'     Flags    : {", ".join(cheating_flags)}')

        overall_pct = round((grand_scored / max(grand_max, 1)) * 100, 1)
        overall_status = '⚠️  INTEGRITY ISSUES DETECTED' if any_cheating else '✅ No Issues'
        print(f'\n  🏆 GRAND TOTAL : {grand_scored} / {grand_max} ({overall_pct}%)')
        print(f'  🔒 INTEGRITY   : {overall_status}')
        print('  ' + '─'*60)

    print('\n' + '='*70)

print_combined_report()
```
